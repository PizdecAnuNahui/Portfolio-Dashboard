```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import time
from sklearn.metrics import mean_squared_error
from statsmodels.tsa.arima.model import ARIMA
from keras.models import Sequential
from keras.layers import Dense, LSTM
from prophet import Prophet
import os
```

```python
# Step 1: Define stock tickers and date range
tickers = ["AAPL", "JPM", "AMZN", "JNJ", "XOM", "CAT", "GOOGL", "NEE", "AMT", "PG"]
start_date = "2019-01-01"
end_date = "2025-03-22"

# Step 2: Download and save each ticker's data
for ticker in tickers:
    print(f"\nüì• Downloading data for {ticker}...")
    try:
        data = yf.download(ticker, start=start_date, end=end_date)
        if not data.empty:
            csv_filename = f"{ticker}_stock_data.csv"
            data.to_csv(csv_filename)
            print(f"‚úÖ Saved {ticker} data to {csv_filename}")
        else:
            print(f"‚ùå No data returned for {ticker}")
    except Exception as e:
        print(f"‚ùå Failed to download {ticker}: {e}")

    # Wait to avoid rate limiting
    time.sleep(5)
```

```python


# ‚úÖ List of CSV files and their labels
tickers = {
    "AAPL": "AAPL_stock_data.csv",
    "JPM": "JPM_stock_data.csv",
    "AMZN": "AMZN_stock_data.csv",
    "JNJ": "JNJ_stock_data.csv",
    "XOM": "XOM_stock_data.csv",
    "CAT": "CAT_stock_data.csv",
    "GOOGL": "GOOGL_stock_data.csv",
    "NEE": "NEE_stock_data.csv",
    "AMT": "AMT_stock_data.csv",
    "PG": "PG_stock_data.csv"
}

summary_data = []

for ticker, file in tickers.items():
    if not os.path.exists(file):
        print(f"‚ö†Ô∏è Skipping {ticker}: File '{file}' not found.")
        continue

    print(f"\nüì• Processing {ticker}...")

    df = pd.read_csv(file, index_col=0, parse_dates=True)

    # Use only 'Close' column and clean
    if 'Close' not in df.columns:
        print(f"‚ùå No 'Close' column in {file}. Skipping.")
        continue

    df = df[['Close']].copy()
    df = df[pd.to_numeric(df['Close'], errors='coerce').notnull()]
    df['Close'] = df['Close'].astype(float)

    if len(df) < 50:
        print(f"‚ùå Not enough data for {ticker}. Skipping.")
        continue

    # Train/Test Split
    train_size = int(len(df) * 0.8)
    train, test = df[:train_size], df[train_size:]

    # Fit ARIMA model
    try:
        model = ARIMA(train, order=(5, 1, 0))
        model_fit = model.fit()
        forecast = model_fit.forecast(steps=len(test))
        forecast = pd.Series(forecast.values, index=test.index)
    except Exception as e:
        print(f"‚ùå ARIMA failed for {ticker}: {e}")
        continue

    # Plot actual vs forecast
    plt.figure(figsize=(10, 4))
    plt.plot(train.index, train, label='Train')
    plt.plot(test.index, test, label='Test')
    plt.plot(forecast.index, forecast, label='Forecast', color='red')
    plt.title(f"ARIMA Forecast vs Actual for {ticker}")
    plt.xlabel("Date")
    plt.ylabel("Close Price")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # RMSE Evaluation
    rmse = np.sqrt(mean_squared_error(test, forecast))
    print(f"üìâ RMSE for ARIMA model on {ticker}: {rmse:.2f}")
    summary_data.append({"Ticker": ticker, "RMSE": round(rmse, 2)})

# ‚úÖ Summary Table
summary_df = pd.DataFrame(summary_data)
print("\nüìä ARIMA Forecast RMSE Summary:")
print(summary_df)
```

```python
# LSTM RMSE results
lstm_rmse = []

# Loop through each stock
for ticker in tickers:
    try:
        print(f"\nüìä Processing {ticker}...")

        # Step 1: Load and clean data
        df = pd.read_csv(f"{ticker}_stock_data.csv", index_col=0, parse_dates=True)
        df = df[['Close']].copy()
        df = df[pd.to_numeric(df['Close'], errors='coerce').notnull()]
        df['Close'] = df['Close'].astype(float)

        # Step 2: Normalize data
        scaler = MinMaxScaler(feature_range=(0, 1))
        scaled_data = scaler.fit_transform(df)

        # Step 3: Create sequences for LSTM
        sequence_length = 60
        X, y = [], []
        for i in range(sequence_length, len(scaled_data)):
            X.append(scaled_data[i - sequence_length:i, 0])
            y.append(scaled_data[i, 0])
        X, y = np.array(X), np.array(y)
        X = np.reshape(X, (X.shape[0], X.shape[1], 1))

        # Step 4: Train/test split
        train_size = int(len(X) * 0.8)
        X_train, y_train = X[:train_size], y[:train_size]
        X_test, y_test = X[train_size:], y[train_size:]

        # Step 5: Build LSTM model
        model = Sequential()
        model.add(LSTM(units=50, return_sequences=False, input_shape=(X.shape[1], 1)))
        model.add(Dense(1))
        model.compile(loss='mean_squared_error', optimizer='adam')

        # Step 6: Train model
        model.fit(X_train, y_train, epochs=10, batch_size=32, verbose=0)

        # Step 7: Forecast
        predictions = model.predict(X_test)
        predictions = scaler.inverse_transform(predictions.reshape(-1, 1))
        y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

        # Step 8: Evaluation
        rmse = np.sqrt(mean_squared_error(y_test_actual, predictions))
        print(f"üìâ RMSE for LSTM on {ticker}: {rmse:.2f}")
        lstm_rmse.append({"Ticker": ticker, "LSTM RMSE": round(rmse, 2)})

        # Step 9: Plot
        plt.figure(figsize=(8, 4))
        plt.plot(y_test_actual, label="Actual")
        plt.plot(predictions, label="LSTM Forecast")
        plt.title(f"LSTM Forecast vs Actual for {ticker}")
        plt.xlabel("Time Step")
        plt.ylabel("Close Price")
        plt.legend()
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"‚ùå Error processing {ticker}: {e}")
        lstm_rmse.append({"Ticker": ticker, "LSTM RMSE": None})

# Step 10: Show Summary
lstm_rmse_df = pd.DataFrame(lstm_rmse)
print("\nüìã LSTM RMSE Summary:")
print(lstm_rmse_df)
```

```python
# Store RMSE results
arima_rmse_results = []
lstm_rmse_results = []

for ticker in tickers:
    print(f"\nüìä Processing {ticker}...")

    # Step 1: Load and clean data
    file = f"{ticker}_stock_data.csv"
    df = pd.read_csv(file, index_col=0, parse_dates=True)
    df = df[['Close']].copy()
    df = df[pd.to_numeric(df['Close'], errors='coerce').notnull()]
    df['Close'] = df['Close'].astype(float)

    # Train-test split (80/20)
    train_size = int(len(df) * 0.8)
    train, test = df[:train_size], df[train_size:]

    # ======== ARIMA ========
    try:
        arima_model = ARIMA(train, order=(5, 1, 0))
        arima_fit = arima_model.fit()
        forecast_arima = arima_fit.forecast(steps=len(test))
        forecast_arima = pd.Series(forecast_arima.values, index=test.index)
        arima_rmse = np.sqrt(mean_squared_error(test, forecast_arima))
    except Exception as e:
        print(f"‚ö†Ô∏è ARIMA failed on {ticker}: {e}")
        arima_rmse = np.nan

    arima_rmse_results.append({"Ticker": ticker, "ARIMA_RMSE": arima_rmse})

    # ======== LSTM ========
    try:
        # Scale the data
        scaler = MinMaxScaler()
        scaled_data = scaler.fit_transform(df[['Close']])

        # Create sequences
        def create_sequences(data, seq_len=10):
            X, y = [], []
            for i in range(len(data) - seq_len):
                X.append(data[i:i + seq_len])
                y.append(data[i + seq_len])
            return np.array(X), np.array(y)

        seq_len = 10
        X, y = create_sequences(scaled_data, seq_len)

        X_train, X_test = X[:train_size - seq_len], X[train_size - seq_len:]
        y_train, y_test = y[:train_size - seq_len], y[train_size - seq_len:]

        # Define and train LSTM model
        model = Sequential()
        model.add(LSTM(50, activation='relu', input_shape=(seq_len, 1)))
        model.add(Dense(1))
        model.compile(optimizer='adam', loss='mse')
        model.fit(X_train, y_train, epochs=10, verbose=0)

        # Predict and invert scale
        y_pred = model.predict(X_test)
        y_pred_inverse = scaler.inverse_transform(y_pred)
        y_test_inverse = scaler.inverse_transform(y_test)

        lstm_rmse = np.sqrt(mean_squared_error(y_test_inverse, y_pred_inverse))
    except Exception as e:
        print(f"‚ö†Ô∏è LSTM failed on {ticker}: {e}")
        lstm_rmse = np.nan

    lstm_rmse_results.append({"Ticker": ticker, "LSTM_RMSE": lstm_rmse})

# ======== Summary Table ========
arima_df = pd.DataFrame(arima_rmse_results)
lstm_df = pd.DataFrame(lstm_rmse_results)
comparison_df = pd.merge(arima_df, lstm_df, on="Ticker")

# Display summary table
print("\nüìâ ARIMA vs LSTM RMSE Summary:")
print(comparison_df)

# ======== Visualization ========
comparison_df.set_index("Ticker").plot(kind='bar', figsize=(10, 5))
plt.title("ARIMA vs LSTM Forecasting RMSE Comparison")
plt.ylabel("RMSE")
plt.grid(True)
plt.tight_layout()
plt.legend(["ARIMA", "LSTM"])
plt.show()
```

```python
# Results list
prophet_rmse_results = []

for ticker in tickers:
    try:
        print(f"\nüìà Processing Prophet model for {ticker}...")

        df = pd.read_csv(f"{ticker}_stock_data.csv", index_col=0, parse_dates=True)
        df = df[['Close']].copy()
        df = df[pd.to_numeric(df['Close'], errors='coerce').notnull()]
        df['Close'] = df['Close'].astype(float)
        df.reset_index(inplace=True)
        df.columns = ['ds', 'y']  # Prophet needs these column names

        # Train-test split
        train_size = int(len(df) * 0.8)
        train = df.iloc[:train_size]
        test = df.iloc[train_size:]

        # Prophet model
        model = Prophet(daily_seasonality=False)
        model.fit(train)

        future = model.make_future_dataframe(periods=len(test), freq='D')
        forecast = model.predict(future)
        forecast_trimmed = forecast[['ds', 'yhat']].iloc[-len(test):].set_index('ds')
        test_trimmed = test.set_index('ds')

        # Align forecast with test set
        common_index = forecast_trimmed.index.intersection(test_trimmed.index)
        y_pred = forecast_trimmed.loc[common_index]['yhat']
        y_true = test_trimmed.loc[common_index]['y']

        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        prophet_rmse_results.append({"Ticker": ticker, "Prophet_RMSE": rmse})
        print(f"‚úÖ RMSE for Prophet on {ticker}: {rmse:.2f}")

        # Optional: plot forecast vs actual
        plt.figure(figsize=(8, 4))
        plt.plot(y_true, label='Actual')
        plt.plot(y_pred, label='Prophet Forecast')
        plt.title(f"Prophet Forecast vs Actual for {ticker}")
        plt.xlabel("Date")
        plt.ylabel("Close Price")
        plt.legend()
        plt.tight_layout()
        plt.show()

    except Exception as e:
        print(f"‚ùå Prophet failed on {ticker}: {e}")
        prophet_rmse_results.append({"Ticker": ticker, "Prophet_RMSE": None})

# ‚úÖ Combine all RMSEs into one summary
prophet_df = pd.DataFrame(prophet_rmse_results)
full_comparison = pd.merge(comparison_df, prophet_df, on="Ticker")

# Display full comparison
print("\nüìä ARIMA vs LSTM vs Prophet RMSE Summary:")
print(full_comparison)

# ‚úÖ Visualize
full_comparison.set_index("Ticker").plot(kind='bar', figsize=(12, 6))
plt.title("ARIMA vs LSTM vs Prophet Forecasting RMSE Comparison")
plt.ylabel("RMSE")
plt.grid(True)
plt.legend(["ARIMA", "LSTM", "Prophet"])
plt.tight_layout()
plt.show()
```

```python
# Propher
prophet_results = []

for ticker in tickers:
    try:
        file_path = f"{ticker}_stock_data.csv"
        df = pd.read_csv(file_path, index_col=0, parse_dates=True)

        # ‚úÖ Prepare for Prophet
        df = df[['Close']].copy()
        df = df[pd.to_numeric(df['Close'], errors='coerce').notnull()]
        df['Close'] = df['Close'].astype(float)
        df = df.reset_index()
        df.columns = ['ds', 'y']  # Prophet expects these exact names

        # ‚úÖ Train/test split
        train_size = int(len(df) * 0.8)
        train_df = df.iloc[:train_size]
        test_df = df.iloc[train_size:]

        # ‚úÖ Fit Prophet model
        model = Prophet(daily_seasonality=True)
        model.fit(train_df)

        # ‚úÖ Forecast on test set dates
        future = test_df[['ds']]
        forecast = model.predict(future)

        # ‚úÖ RMSE calculation
        actual = test_df['y'].values
        predicted = forecast['yhat'].values
        rmse = np.sqrt(mean_squared_error(actual, predicted))

        prophet_results.append({'Ticker': ticker, 'Prophet_RMSE': rmse})

        # ‚úÖ Plot forecast
        plt.figure(figsize=(10, 4))
        plt.plot(actual, label='Actual')
        plt.plot(predicted, label='Prophet Forecast', color='orange')
        plt.title(f'Prophet Forecast vs Actual for {ticker}')
        plt.xlabel("Time Step")
        plt.ylabel("Close Price")
        plt.grid(True)
        plt.legend()
        plt.tight_layout()
        plt.show()

        print(f"‚úÖ Prophet RMSE for {ticker}: {rmse:.2f}")

    except Exception as e:
        print(f"‚ùå Prophet failed on {ticker}: {e}")
        prophet_results.append({'Ticker': ticker, 'Prophet_RMSE': None})
```
